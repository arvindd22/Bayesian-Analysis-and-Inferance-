$\sigma \sim \mathcal{halfnormal}(0.75)$

1.1 - Identify the prior and the likelihood.

1.1

- **Likelihood**: The likelihood function describes the probability of the observed data \(Y\) given the parameters \(\mu\) and \(\sigma\). Here, the likelihood is specified by a normal distribution:


\begin{align*}
Y &\sim N(\mu, \sigma) \\
\end{align*}

- **Priors**: Priors represent our beliefs about the parameters (denoted as \(\Theta\)) before observing any data. In this case, the parameters are \(\mu\) and \(\sigma\).

  Thus, the prior distributions are:


\begin{align*}
\mu &\sim N(0, 2) \\
\sigma &\sim \text{halfnormal}(0.75)
\end{align*}



1.2

The posterior distribution incorporates information from both the likelihood and the priors. As stated in (1.1), this model involves **two** parameters: \(\mu\) and \(\sigma\).

Therefore, the posterior distribution will be a joint distribution over these two parameters.

1.3

In comparison, the coin-flipping problem typically uses a Beta-Binomial model for the likelihood and a Beta distribution for the prior. Specifically, if \(Y\) is the number of heads in \(n\) flips:


\begin{align*}
\text{Likelihood}: & \quad Y \mid p \sim \text{Binomial}(n, p) \\
\text{Prior}: & \quad p \sim \text{Beta}(\alpha, \beta)
\end{align*}

Here, there is only **one** parameter \(p\), which represents the probability of getting heads in a single flip.

Comparing the two models:

1. **Number of Parameters**:

- This normal model involves **two continuous parameters**: \(\mu\) and \(\sigma\).
- The coin-flipping model involves only **one continuous parameter**: \(p\).

2. **Nature of the Priors and Likelihood**:

- This model has priors where \(\mu\) is normally distributed and \(\sigma\) follows a half-normal distribution. The likelihood is also normally distributed.
- The coin-flipping model typically involves a Beta distribution for the prior and a Binomial or Bernoulli distribution for the likelihood.

3. **Complexity of the Posterior Distribution**:

- The posterior distribution in this model will be a joint distribution over two parameters, increasing complexity. The computation of the posterior can become more challenging, and visualizing it involves plotting in two dimensions.

  Posterior Example:

$$
P(\mu, \sigma \mid Y) = \frac{P(Y \mid \mu, \sigma) P(\mu) P(\sigma)}{P(Y)}
$$

- The coin-flipping model's posterior is a function of a single parameter, making analysis and computation simpler. Analytical solutions are often more feasible, and visualizing the posterior is straightforward as it involves plotting in one dimension.

  Posterior Example:

$$
P(p \mid Y) = \frac{P(Y \mid p) P(p)}{P(Y)}
$$

1.4

Bayes' theorem for the given model:

$$
P(\mu, \sigma \mid Y) = \frac{P(Y \mid \mu, \sigma) P(\mu) P(\sigma)}{P(Y)}
$$

Substituting the given distributions:

$$
P(\mu, \sigma \mid Y) = \frac{N(Y \mid \mu, \sigma) \cdot N(\mu \mid 0, 2) \cdot \text{halfnormal}(\sigma \mid 0.75)}{P(Y)}
$$

Also,

$$
P(\mu, \sigma \mid Y) \propto N(Y \mid \mu, \sigma) \cdot N(\mu \mid 0, 2) \cdot \text{halfnormal}(\sigma \mid 0.75)
$$

Letâ€™s suppose that we have two coins; when we toss the first coin, half of the time it lands on tails
 and half of the time on heads. The other coin is a loaded coin that always lands on heads. If we take
 one of the coins at random and get a head, what is the probability that this coin is the unfair one?

```python
'''Given:
Coin 1 (fair): P(Heads) = 0.5, P(Tails) = 0.5
Coin 2 (unfair): P(Heads) = 1, P(Tails) = 0

Our events:
A: The chosen coin is the unfair one (Coin 2)
B: We observe a head

We need to find P(A|B), the probability that the coin is unfair given that we observed a head.

P(A), the probability of choosing the unfair coin.
Since we choose a coin at random, P(A) = 0.5.

P(B|A), the probability of getting a head given that we chose the unfair coin.
P(B|A) = 1, because the unfair coin always lands on heads.

P(B|not A), the probability of getting a head given that we chose the fair coin.
P(B|not A) = 0.5, because the fair coin lands on heads half the time.

P(B), the overall probability of getting a head.
P(B) = P(B|A) * P(A) + P(B|not A) * P(not A)
P(B) = 1 * 0.5 + 0.5 * 0.5 = 0.5 + 0.25 = 0.75

Bayes theorem to calculate P(A|B)
P(A|B) = (P(B|A) * P(A)) / P(B)
P(A|B) = (1 * 0.5) / 0.75
P(A|B) = 0.5 / 0.75
P(A|B) = 2/3 ~ 0.667

Therefore, given that we observed a head, the probability that the coin is the unfair one (Coin 2) is 2/3. '''
```
